```python
import pandas as pd

# Create sample battery degradation data
data = {
    "Cycle": list(range(1, 21)),
    "Capacity": [100, 99, 98, 97, 96, 95, 94, 93, 92, 91,
                 90, 89, 88, 87, 86, 85, 84, 83, 82, 80]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.lineplot(x="Cycle", y="Capacity", data=df, marker="o")
plt.title("Battery Capacity Degradation Over Cycles")
plt.xlabel("Charge Cycles")
plt.ylabel("Battery Capacity (%)")
plt.grid(True)
plt.show()
```
